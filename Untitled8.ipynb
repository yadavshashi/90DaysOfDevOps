````python
# Import necessary libraries
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

(a) Tossing a coin 10,000 times

````python
def coin_toss_simulation(trials=10000):
    heads = 0
    tails = 0
    for _ in range(trials):
        if random.choice(['H', 'T']) == 'H':
            heads += 1
        else:
            tails += 1
    prob_heads = heads / trials
    prob_tails = tails / trials
    print(f"Probability of Heads: {prob_heads:.4f}")
    print(f"Probability of Tails: {prob_tails:.4f}")

coin_toss_simulation()
````

(b) Rolling two dice to get a sum of 7

````python
def dice_sum_seven(trials=10000):
    count = 0
    for _ in range(trials):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        if die1 + die2 == 7:
            count += 1
    prob_sum_7 = count / trials
    print(f"Probability of getting sum 7: {prob_sum_7:.4f}")

dice_sum_seven()
````

2️⃣ At Least One '6' in 10 Rolls

````python
def prob_at_least_one_six(trials=10000):
    success = 0
    for _ in range(trials):
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:
            success += 1
    probability = success / trials
    print(f"Estimated probability of at least one '6' in 10 rolls: {probability:.4f}")

prob_at_least_one_six()
````

3️⃣ Conditional Probability and Bayes' Theorem

````python
def conditional_prob_bayes(trials=1000):
    colors = ['R'] * 5 + ['G'] * 7 + ['B'] * 8
    prev_ball = None
    count_prev_blue = 0
    red_given_blue = 0

    for _ in range(trials):
        ball = random.choice(colors)
        if prev_ball == 'B':
            count_prev_blue += 1
            if ball == 'R':
                red_given_blue += 1
        prev_ball = ball

    if count_prev_blue == 0:
        print("No blue balls were drawn previously.")
    else:
        conditional_prob = red_given_blue / count_prev_blue
        print(f"Conditional probability P(Red | Prev Blue): {conditional_prob:.4f}")

conditional_prob_bayes()
````

4️⃣ Discrete Random Variable Sampling

````python
def discrete_random_var():
    X = [1, 2, 3]
    probs = [0.25, 0.35, 0.4]
    sample = np.random.choice(X, size=1000, p=probs)
    mean = np.mean(sample)
    var = np.var(sample)
    std_dev = np.std(sample)
    print(f"Empirical Mean: {mean:.4f}")
    print(f"Empirical Variance: {var:.4f}")
    print(f"Empirical Std Dev: {std_dev:.4f}")

discrete_random_var()
````

5️⃣ Exponential Distribution Simulation & Visualization

````python
def exponential_distribution_simulation():
    mean_value = 5
    samples = np.random.exponential(scale=mean_value, size=2000)

    plt.figure(figsize=(10, 6))
    sns.histplot(samples, bins=30, kde=True, stat='density', color='skyblue', edgecolor='black')
    plt.title('Exponential Distribution (Mean=5)')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()

exponential_distribution_simulation()
````

````python
def central_limit_theorem_sim():
    population = np.random.uniform(0, 1, 10000)
    sample_means = []
    for _ in range(1000):
        sample = np.random.choice(population, size=30)
        sample_means.append(np.mean(sample))

    plt.figure(figsize=(14, 6))

    # Plot 1: Original Uniform Distribution
    plt.subplot(1, 2, 1)
    sns.histplot(population, bins=30, color='green', edgecolor='black')
    plt.title('Uniform Distribution')

    # Plot 2: Distribution of Sample Means
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, bins=30, color='purple', kde=True, edgecolor='black')
    plt.title('Distribution of Sample Means (CLT)')

    plt.show()

central_limit_theorem_sim()
````
